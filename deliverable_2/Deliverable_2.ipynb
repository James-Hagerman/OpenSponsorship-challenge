### What is the ratio of brands that published campaigns vs. did not publish any campaign within each plan level? PL0 = .8 to 1, PL1 = 3 to 1, PL2 = 0, PL3 = 12 to 1, PL9 = 2.6 to 1

```python
query = '''
          SELECT planLevel,
              campaignsPublished,
              campaignsNotPublished,
              campaignsPublished * 1.0 / campaignsNotPublished AS ratioOfPublishedAndNonPublishedBrands
          FROM (SELECT
          users.planLevel,
          COUNT(CASE WHEN campaigns.published IS NOT NULL THEN 1 END) AS campaignsPublished,
          COUNT(CASE WHEN campaigns.published IS NULL THEN 1 END) AS campaignsNotPublished
          FROM campaigns
          INNER JOIN users
          ON campaigns.user_id = users.user_id
          GROUP BY users.planLevel)
  '''
ratio_df = pd.read_sql(query, conn)
ratio_df.head()
```

```python
# Setting planLevel 9 to 4 to make the barchart look better
ratio_df['planLevel'][4] = 4

# Using ggplot style 
plt.style.use('ggplot')


fig, ax = plt.subplots()
fig.set_facecolor("#F0F0F0")


bar_width = 0.35

# Plotting bars for campaigns published
ax.bar(ratio_df['planLevel'] - bar_width/1.9, ratio_df['campaignsPublished'], width=bar_width, label='campaignsPublished')

# Plotting bars for campaigns not published 
ax.bar(ratio_df['planLevel'] + bar_width/1.9, ratio_df['campaignsNotPublished'], width=bar_width, label='campaignsNotPublished')

plt.title('Ratio of Campaigns Published to Campaigns Not Published', 
             fontsize = 12, fontweight = 'bold', x= .45, y = 1.03)

# Setting x tick labels
ax.set_xticklabels([0, 1, 2, 3, 4, 9], fontsize = '8')

ax.set_xlabel('planLevel')
ax.set_ylabel('Number of Campaigns')

plt.legend()

# Add count labels on top of the bars
for p in ax.patches:
    height = int(p.get_height())
    if height > 0:
        ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='bottom', fontsize=10, color='black')

plt.show()
```

### How many brands continued to create campaigns after their first published one? - 15 brands published more than one campaign.

```python
query = '''SELECT COUNT(numOfCampaigns) AS multipleCampaignBrands
           FROM (
               SELECT COUNT(DISTINCT campaign_id) as numOfCampaigns
               FROM campaigns
               WHERE published IS NOT NULL 
               GROUP BY user_id
               ORDER BY COUNT(DISTINCT campaign_id) DESC 
            ) 
            WHERE numOfCampaigns > 1
           
           '''
num_published_df = pd.read_sql(query, conn)
num_published_df.head()
```

### Which brands have the highest paying campaigns?

```python
# Operating under the assumption that "highest paying" refers to total budget 
query = '''SELECT user_id, payment
           FROM campaigns
           '''
budget_df = pd.read_sql(query, conn)
budget_df.head()
```

```python
# Function to extract the budget from json 
def extract_budget(json_str):

    # Converting to a python dictionary 
    dict = ast.literal_eval(json_str)

    # Retrieving the different budget values if they are available 
    cash_budget = dict.get("cash", {}).get("budget") if dict.get("cash", {}).get('available') else 0
    product_budget = dict.get("product", {}).get("budget") if dict.get("product", {}).get('available') else 0
    royalty_budget = dict.get("royalty", {}).get("budget") if dict.get("royalty", {}).get('available') else 0
    equity_budget = dict.get("equity", {}).get("budget") if dict.get("equity", {}).get('available') else 0

    # Turning budgets into integers and handling empty cases
    cash_budget = int(cash_budget) if cash_budget else 0 
    product_budget = int(product_budget) if product_budget else 0 
    royalty_budget = int(royalty_budget) if royalty_budget else 0 
    equity_budget = int(equity_budget) if equity_budget else 0 

    # Summing the budgets 
    total_budget = cash_budget + product_budget + royalty_budget + equity_budget
    return total_budget


# Applying function to the data 
budget_df['totalBudget'] = budget_df['payment'].apply(extract_budget)
# Dropping payment column 
budget_df = budget_df.drop(['payment'], axis = 1)
budget_df.head()
```

```python
# Selecting the 10 largets budget users 
budget_df=budget_df.groupby('user_id').mean()
budget_df = budget_df.nlargest(10, 'totalBudget')
budget_df = budget_df.reset_index()
```

```python
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(4, 8))

fig.set_facecolor("#F0F0F0")

# Plotting the data
sns.barplot(data=budget_df, x= 'totalBudget', y = 'user_id', orient = 'h')

# Adjusting the limits of the chart to fit the data
plt.ylim(-1, 10)  
plt.xlim(0, 130000)

plt.xticks(rotation=45)

ax.set_title('Top 10 Highest Paying Brands', 
             fontsize = 12, fontweight = 'bold')
plt.show()
```

### How long does it take for new brands to publish a campaign? - On average it takes 54.38 days for a new brand to publish their first campaign

```python
query = '''
          SELECT user_id, MIN(dateDiff) AS daysToPublishFromAccountCreation
          FROM (SELECT 
              users.user_id,
              JULIANDAY(campaigns.published) - JULIANDAY(users.created) AS dateDiff
          FROM campaigns
          INNER JOIN users
          ON campaigns.user_id = users.user_id
          WHERE dateDiff IS NOT NULL) 
          GROUP BY user_id
          
        '''
time_to_publish_df = pd.read_sql(query, conn)
# Printing the average time it takes to publish the first campaign 
print("On average it takes new brands " + str(round(time_to_publish_df['daysToPublishFromAccountCreation'].mean(), 2)) + " days to publish their first campaign.")
time_to_publish_df.head()
```

```python
plt.style.use('ggplot')


fig, ax = plt.subplots(figsize=(20,4))
fig.set_facecolor("#F0F0F0")

sns.barplot(data=time_to_publish_df, x='user_id', y = 'daysToPublishFromAccountCreation')

ax.set_xlim(-.5, 88)  
ax.set_ylim(0, 450)

ax.set_xlabel('User ID')
ax.set_ylabel('Days to First Publish')

plt.xticks(rotation=90, fontsize = 10)

# PLotting the average line 
plt.axhline(y=time_to_publish_df['daysToPublishFromAccountCreation'].mean(), color='maroon', linestyle='-', linewidth=2, label='Avg')


ax.set_title('How Long it Takes For Brands to Publish From Account Creation', 
             fontsize = 12, fontweight = 'bold')

for p in ax.patches:
    height = int(p.get_height())
    if height > 0:
        ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='bottom', fontsize=10, color='black')

plt.show()
```
